## 불쾌지수
80이상 : 매우높음 - 전원 불쾌감

80~75 : 높음 - 불쾌감을 나타내기 시작

75~68 : 보통 - 50%정도 불쾌감을 느낌

68미만 : 낮음 - 전원 쾌적함을 느낌


```python
# 불쾌지수 분류 함수
```

```python
def make_discomfort(df):
  df['쾌적'] = df['불쾌지수'].apply(lambda x: 1 if  (x < 75) else 0)
  df['불쾌'] = df['불쾌지수'].apply(lambda x: 1 if (75 <= x) else 0)

  return df
```

```python
new_lun_df = make_discomfort(lun_df)
new_eve_df = make_discomfort(eve_df)
```

```python
new_lun_df.head(600).tail(5)
```

## 폭염

- 체감온도 33도 이상

```python
def make_temper(df):
  # df['폭염'] = df['체감온도'].apply(lambda x: 1 if x >= 33.0 else 0)
  # df['한파'] = df['체감온도'].apply(lambda x: 1 if x < (-10.0) # 한파에 대해서
  df['저온'] = df['체감온도'].apply(lambda x: 1 if x < 0.0 else 0)
  df['중온'] = df['체감온도'].apply(lambda x: 1 if (0.0 <= x) & (x < 20.0) else 0)
  df['고온'] = df['체감온도'].apply(lambda x: 1 if 20.0 < x else 0)
  return df
```

```python
new_lun_df = make_temper(new_lun_df)
new_eve_df = make_temper(new_eve_df)
```
